```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

df = pd.read_excel("Ekrem Proje.xlsx")

df["aldigimiz_fiyat"] = df["kg/eur"]
df["piyasa_fiyati"] = df["kg/eur  borsa fiyat"]

df["fiyat_farki"] = df["aldigimiz_fiyat"] - df["piyasa_fiyati"]
df["fiyat_orani"] = df["aldigimiz_fiyat"] / df["piyasa_fiyati"]

model = IsolationForest(contamination=0.1, random_state=42)
df["anomaly"] = model.fit_predict(df[["fiyat_orani"]])

df["durum"] = df["anomaly"].map({-1: "Anomali", 1: "Normal"})

print(df[["Satıcı", "Malzeme", "aldigimiz_fiyat", "piyasa_fiyati", "fiyat_orani", "durum"]])

plt.figure(figsize=(10, 5))
plt.scatter(df.index, df["fiyat_orani"], c=(df["anomaly"] == -1), cmap="coolwarm", label="Anomaliler")
plt.axhline(y=1, color='gray', linestyle='--', label="Piyasa ile aynı")
plt.title("Fiyat Oranına Göre Anomali Tespiti")
plt.xlabel("İşlem")
plt.ylabel("Fiyat Oranı (Aldığımız / Piyasa)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from fpdf import FPDF
import tempfile

def safe_str(text):
    return str(text).encode("latin-1", "replace").decode("latin-1")

df = pd.read_excel("Ekrem Proje2.xlsx")

df["aldigimiz_fiyat"] = df["kg/eur"]
df["piyasa_fiyati"] = df["kg/eur  borsa fiyat"]
df["fiyat_orani"] = df["aldigimiz_fiyat"] / df["piyasa_fiyati"]

z_scores = (df["fiyat_orani"] - df["fiyat_orani"].mean()) / df["fiyat_orani"].std()
df["zscore_flag"] = z_scores.abs() > 3

Q1 = df["fiyat_orani"].quantile(0.25)
Q3 = df["fiyat_orani"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df["iqr_flag"] = (df["fiyat_orani"] < lower_bound) | (df["fiyat_orani"] > upper_bound)

X_scaled = StandardScaler().fit_transform(df[["fiyat_orani"]])
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
df["dbscan_flag"] = dbscan_labels == -1

df["votes"] = df[["zscore_flag", "iqr_flag", "dbscan_flag"]].sum(axis=1)
df["durum"] = df["votes"].apply(lambda x: "Anomali" if x >= 2 else "Normal")

df["Belge tarihi"] = pd.to_datetime(df["Belge tarihi"]).dt.date

anomaliler = df[df["durum"] == "Anomali"]
overpriced_anomaliler = anomaliler[anomaliler["aldigimiz_fiyat"] > anomaliler["piyasa_fiyati"]]

plt.figure(figsize=(14, 6))
colors = df["durum"].map({"Anomali": "red", "Normal": "blue"})
plt.scatter(df.index, df["fiyat_orani"], c=colors)

for i, row in anomaliler.iterrows():
    plt.text(i, row["fiyat_orani"] + 0.01, f"{row['Belge tarihi']} | {row['Satıcı']} | {row['Miktar']}", fontsize=7, rotation=45)

plt.axhline(y=1, color='gray', linestyle='--', label="Piyasa ile aynı")
plt.title("Majority Voting ile Anomali Tespiti (Z-Score, IQR, DBSCAN)")
plt.xlabel("İşlem Sırası")
plt.ylabel("Fiyat Oranı (Aldığımız / Piyasa)")
plt.legend()
plt.tight_layout()

with tempfile.NamedTemporaryFile(delete=False, suffix=".png") as tmpfile1:
    grafik1_path = tmpfile1.name
    plt.savefig(grafik1_path, dpi=300, bbox_inches="tight")
plt.show()

plt.figure(figsize=(14, 6))
df_sorted = df.sort_values("Belge tarihi")
plt.plot(df_sorted["Belge tarihi"], df_sorted["kg/eur"], label="Aldığımız Fiyat (kg/eur)", marker="o")
plt.plot(df_sorted["Belge tarihi"], df_sorted["kg/eur  borsa fiyat"], label="Piyasa Fiyatı (kg/eur)", marker="x")

etiketlencekler = df_sorted[
    (df_sorted["durum"] == "Anomali") &
    (df_sorted["aldigimiz_fiyat"] > df_sorted["piyasa_fiyati"])
]

for i, row in etiketlencekler.iterrows():
    plt.annotate(
        f"{row['Belge tarihi']}\nMiktar: {row['Miktar']:,}",
        xy=(row["Belge tarihi"], row["kg/eur"]),
        xytext=(0, 15),
        textcoords="offset points",
        ha="center",
        fontsize=8,
        color="red",
        arrowprops=dict(arrowstyle="->", color="gray", lw=1.0)
    )

plt.title("Günlük Aldığımız Fiyat vs Piyasa Fiyatı (Sadece Overpriced Anomaliler Gösterildi)")
plt.xlabel("Tarih")
plt.ylabel("Fiyat (kg/EUR)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

with tempfile.NamedTemporaryFile(delete=False, suffix=".png") as tmpfile2:
    grafik2_path = tmpfile2.name
    plt.savefig(grafik2_path, dpi=300, bbox_inches="tight")
plt.show()

with tempfile.NamedTemporaryFile(delete=False, suffix=".png") as tmpfile3:
    plt.figure(figsize=(14, 6))
    plt.plot(df_sorted["Belge tarihi"], df_sorted["aldigimiz_fiyat"], label="Aldığımız Fiyat", linewidth=2)
    plt.plot(df_sorted["Belge tarihi"], df_sorted["piyasa_fiyati"], label="Piyasa Fiyatı", linewidth=2)
    plt.title("Fiyatların Tarihsel Gelişimi")
    plt.xlabel("Tarih")
    plt.ylabel("Fiyat (kg/EUR)")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True)
    grafik3_path = tmpfile3.name
    plt.savefig(grafik3_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

excel_dosya_adi = "anomaliler.xlsx"
with pd.ExcelWriter(excel_dosya_adi, engine="openpyxl", mode="w") as writer:
    anomaliler.to_excel(writer, sheet_name="tum_anomaliler", index=False)
    overpriced_anomaliler.to_excel(writer, sheet_name="overpriced_anomaliler", index=False)

print(f"Anomaliler '{excel_dosya_adi}' dosyasına yazıldı.")

# PDF Rapor
class PDF(FPDF):
    def __init__(self):
        super().__init__(orientation='L', unit='mm', format='A4')

    def header(self):
        self.set_font("Arial", "B", 12)
        self.multi_cell(0, 10, safe_str("Torun Metal - Son 5 Yila Ait CW614N/CW617N\nHammaddesi Satin Alim Fiyatlarindaki Anomaliler"), align="C")
        self.ln(5)

    def footer(self):
        self.set_y(-15)
        self.set_font("Arial", "I", 6)
        footer_text = f"Sayfa {self.page_no()}  |  Ekrem Kirdemir"
        self.cell(0, 10, safe_str(footer_text), align="C")

pdf = PDF()
pdf.add_page()
pdf.set_font("Arial", size=8)

columns = ["Belge tarihi", "Üretim yeri", "Ad 1", "Satici", "Malzeme", "Miktar", "Fiyat",
           "aldigimiz_fiyat", "piyasa_fiyati", "fiyat_orani",
           "zscore_flag", "iqr_flag", "dbscan_flag"]

mevcut_sutunlar = set(overpriced_anomaliler.columns)
columns = [col for col in columns if col in mevcut_sutunlar]
col_widths = []
for col in columns:
    max_len = overpriced_anomaliler[col].astype(str).apply(len).max()
    width = min(35, max(15, max_len * 2.5))
    if col == "Ad 1":
        width += 20
    col_widths.append(width)

for col, w in zip(columns, col_widths):
    pdf.cell(w, 8, safe_str(col), border=1, align='C')
pdf.ln()

for _, row in overpriced_anomaliler.iterrows():
    for col, w in zip(columns, col_widths):
        val = row[col]
        if col in ["aldigimiz_fiyat", "fiyat_orani"]:
            val = f"{val:.2f}"
        elif col == "Belge tarihi":
            val = row[col].strftime("%Y-%m-%d")
        val = safe_str(val)
        pdf.cell(w, 8, val, border=1)
    pdf.ln()

total_loss = ((overpriced_anomaliler["aldigimiz_fiyat"] - overpriced_anomaliler["piyasa_fiyati"]) * overpriced_anomaliler["Miktar"]).sum()
total_qty = overpriced_anomaliler["Miktar"].sum()
avg_ratio = overpriced_anomaliler["fiyat_orani"].mean()
total_rows = overpriced_anomaliler.shape[0]

pdf.ln(5)
pdf.set_font("Arial", "B", 7)
pdf.cell(0, 8, safe_str(f"Toplam Zarar: {total_loss:,.2f} EUR"), ln=True)
pdf.cell(0, 8, safe_str(f"Toplam Miktar: {total_qty:,.0f}"), ln=True)
pdf.cell(0, 8, safe_str(f"Ortalama Fiyat Orani: {avg_ratio:.2f}"), ln=True)
pdf.cell(0, 8, safe_str(f"Anomali Satir Sayisi: {total_rows}"), ln=True)

pdf.add_page()
pdf.image(grafik1_path, x=10, y=40, w=270, h=100)

pdf.add_page()
pdf.image(grafik2_path, x=10, y=40, w=270, h=100)

pdf.add_page()
pdf.image(grafik3_path, x=10, y=40, w=270, h=100)

try:
    pdf.output("overpriced_anomali_raporu.pdf")
    print("PDF oluşturuldu: overpriced_anomali_raporu.pdf")
except Exception as e:
    print("PDF oluşturulamadı:", str(e))
```
